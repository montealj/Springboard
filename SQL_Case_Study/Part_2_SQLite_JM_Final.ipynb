```python
#Import Packages
from sqlalchemy import create_engine
import os
import pandas as pd
```

```python
os.getcwd()
```

```python
#Create Engine

engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

```python
Q10 = pd.read_sql_query("SELECT f.name AS Facility, SUM((CASE WHEN b.memid=0 THEN f.guestcost * b.slots ELSE f.membercost * b.slots END)) AS total_rev FROM Bookings AS b INNER JOIN Facilities AS f ON b.facid = f.facid GROUP BY Facility HAVING 1000> total_rev ORDER BY total_rev", engine)
```

```python
Q10
```

```python
Q11 = pd.read_sql_query("SELECT m1.surname ||' '||m1.firstname AS Member, m2.surname||' '||m2.firstname AS Recommended_By FROM Members AS m1 INNER JOIN Members AS m2 ON m1.recommendedby = m2.memid WHERE m1.recommendedby <> 0 ORDER BY m1.surname", engine)
```

```python
Q11
```

```python
#Alternative if wasnt first, last then ordered by last of recommender
aQ11 = pd.read_sql_query("SELECT m1.firstname||' '||m1.surname AS Member, m2.firstname||' '||m2.surname AS Recommended_By FROM Members AS m1 INNER JOIN Members AS m2 ON m1.recommendedby = m2.memid WHERE m1.recommendedby <> 0 ORDER BY m1.surname", engine)
```

```python
Q12 = pd.read_sql_query("SELECT f.name AS Facility, b.memid, SUM(b.slots) AS Number_of_slots_booked  FROM Bookings AS b INNER JOIN Facilities AS f ON b.facid = f.facid WHERE b.memid <> 0 GROUP BY b.memid, Facility", engine)
```

```python
Q12
```

```python
Q13 = pd.read_sql_query("SELECT facid, strftime('%m', starttime) AS Month, SUM(slots) AS Number_of_slots_booked FROM Bookings WHERE memid <> 0 GROUP BY facid, Month", engine)
```
